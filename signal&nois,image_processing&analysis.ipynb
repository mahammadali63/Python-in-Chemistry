```
import pandas as pd
import numpy as np
from scipy.signal import find_peaks
import matplotlib.pyplot as plt
data=pd.read_csv("/content/data_SciCompforChem/CV_K3Fe(CN)6.csv", header=None)
data.columns=['Potential (V)', 'Current (A)']
potential=data['Potential (V)']
current=data['Current (A)']
peaks_max,_=find_peaks(current)
peaks_min,_=find_peaks(-current)
plt.figure(figsize=(15,10))
plt.plot(potential, current,label='CV curve', color='blue')
plt.plot(potential[peaks_max],current[peaks_max],'go',label='Maxima')
plt.plot(potential[peaks_min],current[peaks_min],'r^',label='Minima')
plt.xlabel('Potential (V)')
plt.ylabel('Current (A)')
plt.title('Cyclic Voltammogram of K3Fe(CN)6')
plt.legend()
plt.grid(True)
plt.show()
max_points=pd.DataFrame({
    'Potential (V)':potential[peaks_max],
    'Current (A)':current[peaks_max]
})
min_points=pd.DataFrame({
    'Potential (V)':potential[peaks_min],
    'Current (A)':current[peaks_min]
})
print("Maximum Points:\n",max_points)
print("\nMinimum Points:\n",min_points)
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
conc=np.array([0.10,0.16,0.20,0.25,0.41,0.55])
rate=np.array([0.0034,0.0087,0.014,0.021,0.057,0.10])
def rate_law(A,k,n):
  return A*k**n
popt,pcov=curve_fit(rate_law,conc,rate,p0=[1,1])
k_fit,n_fit=popt
print(f"fitted k={k_fit:.4f}")
print(f"fitted n={n_fit:.4f}")
A_fit=np.linspace(0.0,0.6,100)
rate_fit=rate_law(A_fit,k_fit,n_fit)
plt.figure(figsize=(10,6))
plt.plot(conc,rate,'o',label='data')
plt.plot(A_fit,rate_fit,label=f'fit:k={k_fit:.3f},n={n_fit:.3f}')
plt.xlabel('[A]')
plt.ylabel('rate')
plt.title('Rate vs Concentration')
plt.legend()
plt.grid(True)
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
r=np.array([1,5,9,13,17])
psi=np.array([0.21,-0.087,-0.027,-0.0058,-0.00108])
interp_func=interp1d(r,psi,kind='cubic')
r_smooth=np.linspace(r.min(),r.max(),50000)
psi_smooth=interp_func(r_smooth)
plt.figure(figsize=(10,6))
plt.plot(r_smooth,psi_smooth,'b-',label='2s radial wave function')
plt.plot(r,psi,'ro',label='data point') #origins
plt.xlabel('r(bohr)')
plt.ylabel('(r)')
plt.title('Radial Wave Function of H')
plt.legend()
plt.grid(True)
plt.show()

```

```
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
from skimage.feature import blob_log
image=io.imread("spots.png")
gray=color.rgb2gray(image)
blobs=blob_log(gray,
                 min_sigma=3,
                 max_sigma=20,
                 num_sigma=10,
                 threshold=0.1)
blobs[:,2]=blobs[:,2]*np.sqrt(2)
print("numbers of founded spots:",len(blobs))
plt.figure(figsize=(8,8))
plt.imshow(gray, cmap='gray')
for blob in blobs:
    y,x,r=blob
    plt.plot(x, y, 'rx')#red X
plt.title("founded spots")
plt.axis('off')
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
from skimage.filters.rank import entropy
from skimage.morphology import disk
from skimage.util import img_as_ubyte
image=io.imread("test_tube_altered.png")
gray=color.rgb2gray(image)
gray_ubyte=img_as_ubyte(gray)
entropy_map=entropy(gray_ubyte,disk(5))
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.imshow(gray,cmap='gray')
plt.title("Original image")
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(entropy_map, cmap='inferno')
plt.title("Entropy Map")
plt.colorbar()
plt.axis('off')
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from skimage import io,color,img_as_float
img_png=io.imread("nmr.png")
img_jpg=io.imread("nmr.jpg")
if img_png.shape[-1]==4:
    img_png=img_png[:,:,:3]
if img_jpg.shape[-1]==4:
    img_jpg=img_jpg[:,:,:3]
img_png_gray=color.rgb2gray(img_png)
img_jpg_gray=color.rgb2gray(img_jpg)
img_png_f=img_as_float(img_png_gray)
img_jpg_f=img_as_float(img_jpg_gray)
diff=np.abs(img_png_f-img_jpg_f)
plt.figure(figsize=(12,5))
plt.subplot(1,3,1)
plt.imshow(img_png_f,cmap='gray')
plt.title("NMR PNG (Lossless)")
plt.axis('off')
plt.subplot(1,3,2)
plt.imshow(img_jpg_f,cmap='gray')
plt.title("NMR JPG (Lossy)")
plt.axis('off')
plt.subplot(1,3,3)
plt.imshow(diff,cmap='hot')
plt.title("difference (PNG-JPG)")
plt.axis('off')
plt.show()
```
