```
def classify_compound(formula):
    #Remove parentheses
    formula = formula.replace("(", "").replace(")", "")

    if formula.startswith("H"):
        return "acid"
    elif "OH" in formula:
        return "base"
    else:
        return "neutral"

#List of compounds
compounds = ["HCl","NaOH","KCl", "H2SO4", "Ca(OH)2",
             "KOH", "HNO3", "Na2SO4", "KNO3", "Mg(OH)2",
             "HCO2H", "NaBr"]

#Loop compounds and classify
for cmpd in compounds:
    classification = classify_compound(cmpd)
    print(f"{cmpd}: {classification}")

```

```
#as we know from chemisrty, cation is p>e,anion is e>p< and nutron is p=e
def determine_charge(protons, electrons):
      if protons>electrons:
          return "cationic"
      elif electrons>protons:
         return "anionic"
      else:
         return "neutral"

protons = int(input("Enter number of protons: "))
electrons = int(input("enter the number of electrons: "))
charge = determine_charge(protons, electrons)
print(f"the compound is {charge}.")
```
